```python
import PyPDF2
import re

pdf_path = "/Users/a1-6/Project-helsinki/project_30years-butterfly/paper/2023.01.11.523549v1.full.pdf"

def remove_line_numbers(text):
    # 假设行号的格式为一个空格开头，后面跟纯数字，然后两个空格
    cleaned_text = re.sub(r'^\s*\d+\s*\s$', '', text, flags=re.MULTILINE)
    return cleaned_text

def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            num_pages = len(reader.pages)
            for page_num in range(num_pages):
                page = reader.pages[page_num]
                page_text = page.extract_text()
                if page_text:
                    page_text = remove_line_numbers(page_text)
                    text += page_text + "\n"
            return text
    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
        return None

# 使用示例
text = extract_text_from_pdf(pdf_path)
if text:
    print(text)
```

```python
import fitz  # PyMuPDF
import re

def extract_abstract_from_pdf(pdf_path):
    # Open the PDF document
    document = fitz.open(pdf_path)
    text = ""
    
    # Iterate through each page to find the abstract
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text += page.get_text("text") + "\n"
    
    # Use regex to find the abstract section
    match = re.search(r"Abstract\n([\s\S]*?)(Introduction|INTRODUCTION|1\. Introduction|2\. Methods|Materials and Methods|METHODS)", text, re.DOTALL | re.IGNORECASE)
    if match:
        abstract_text = match.group(1).strip()
        
        # Remove line numbers and excessive newlines
        abstract_text = re.sub(r'\n\s*\d+\s*', ' ', abstract_text)  # Remove line numbers
        abstract_text = re.sub(r'\s+', ' ', abstract_text)  # Replace multiple spaces/newlines with a single space
        
        # Remove any remaining artifacts at the end of the abstract
        abstract_text = re.sub(r'(Word count:.*|CC-BY.*|doi:.*|bioRxiv preprint.*| \d+ |^\d+ )', '', abstract_text).strip()
        abstract_text = re.sub(r'^\d+\s*', '', abstract_text)  # Remove leading numbers
        abstract_text = re.sub(r'\s+\d+\s*$', '', abstract_text)  # Remove trailing numbers
        
        return abstract_text
    
    return ""

# Path to the PDF file
pdf_path = "/Users/a1-6/Project-helsinki/project_30years-butterfly/paper/li-et-al-2024-dual-stressors-of-infection-and-warming-can-destabilize-host-microbiomes.pdf"
abstract = extract_abstract_from_pdf(pdf_path)
print("Extracted Abstract:\n", abstract)
```
